In this notebook, we will implement the gradient descent using PyTorch.

We generate a noisy dataset:

````python
import numpy as np

np.random.seed(0)
n_samples = 1000
X = np.random.rand(n_samples)*10
X = np.expand_dims(X, axis=-1)
noise = np.random.randn(n_samples)*4.0
noise = np.expand_dims(noise, axis=-1)
ground_truth = 5 * X + 3
y = ground_truth + noise
````

X and y are the features and the label, respectively. We can look at the format of the data:

````python
print(type(X))
print(type(y))

print(X.shape)
print(y.shape)

print(X.dtype)
print(y.dtype)
````

As you can see, X and y are numpy arrays. We generated 1000 samples, with one feature per sample, and y are the corresponding labels for each sample.

## Performing linear regression with backpropagation

In this notebook we are going to:
 - Implement the gradient descent using pytorch for the calculations instead of numpy. To do this, we will do the following:
   - Import your X, y, numpy tensors into pytorch using `torch.as_tensor(numpy_array)`. For the weights, we will create them using numpy , then convert them to pytorch with `weights = torch.tensor(weights, requires_grad=True)`. This tells pytorch to calculate gradients for this tensor.
   - After calculating the loss, we do not need to calculate the gradient manually. We can use `loss.backward()` if loss is the variable that contains the final loss value. Then, you can access the gradient using `weights.grad`.
   - When updating the weights with `weights -= learning_rate * weights.grad`, wrap it with a block of
   ```python
   with torch.no_grad():
        weights -= learning_rate * weights.grad
   ```
   This tells pytorch to not calculate a gradient for the `-=` operation that updates the weight tensor (which would cause an error).
   - When coding the loop, we need to make sure to reset the gradient calculation after or before each update using `weights.grad.zero_()`. This ensures that the gradients do not accumulate across multiple loop iterations.


## Solution

We add an additional column for the bias to the dataset:

````python
X_ = np.c_[X, np.ones((X.shape[0]))]
````

We initialize a weight matrix with random values:

````python
import torch
num_samples = X_.shape[0]
num_features = X_.shape[1]
X = torch.as_tensor(X_)
y = torch.as_tensor(y)
````

````python
def generate_weights(num_features):
    np.random.seed(42) # this is just so that we generate the same weights everytime we run this code. 
    return (np.random.rand(num_features).reshape(-1, 1) * 2 - 1) / np.sqrt(num_features)

weights = generate_weights(num_features)
print(weights.shape)
weights = torch.tensor(weights, requires_grad=True)
````

````python
predictions = X @ weights


loss = torch.mean((y-predictions)**2)
print(loss)
````

````python
loss.backward()
print(weights.grad)
````

````python
it = 1000
lr = 0.01 #learning rate


for i in range(it):
    predictions = X @ weights
    loss = torch.mean((y-predictions)**2)
    loss.backward()
    with torch.no_grad():
        weights -= lr * weights.grad
    if i % 100 == 0:
        print(loss)
    weights.grad.zero_()
````

````python
print(predictions.shape)
print(y.shape)
print(X.shape)
````

````python
plt.figure()
plt.scatter(X[:,0], y, marker="x")
with torch.no_grad():
    plt.scatter(X[:,0], predictions, marker = "o")
plt.show()
````
